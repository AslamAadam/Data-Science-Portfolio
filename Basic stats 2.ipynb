```python
import pandas as pd
import numpy as np
import math
import scipy.stats as stats
```

```python
data=np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
data = pd.Series(data)
data
```

```python
data.describe()
```

```python
#calculating mean
mean=data.mean()
print(f"Mean of the data is : {mean:.4f}")
```

# a) Build 99% Confidence Interval Using Sample Standard Deviation

```python
std=data.std(ddof=1)
std
#delta degree of freedom [1] for unbaised estimator of variance of population
```

```python
#taking length of our sample data
n=len(data)
print(f"Length of our sample data is : {n}")
```

```python
#creating 99% confidence_level & corresponding alpha value
confidence_interval=0.99
alpha=1-confidence_interval
print(f"Alpha value is : {alpha:.4f}")
```

```python
#creating degree of freedom
dof=n-1
print(f"Degree of freedom is : {dof}")
```

```python
#creating critical value using scipy library
critical_value=stats.t.ppf(1-alpha/2,dof)
print(f"Critical value is : {critical_value:.4f}")
```

```python
#creating marginal_error
margin_error=critical_value*std/math.sqrt(n)
print(f"Marginal error is : {margin_error:.4f}")
```

```python
#calculating confidence interval
lower_bound=mean-margin_error
upper_bound=mean+margin_error
```

```python
print(f"The {confidence_interval*100}% confidence interval for the population mean is :- ")
print(f"lower bound : {lower_bound:.4f}")
print(f"upper Bound : {upper_bound:.4f}")
```

# b. Build 99% Confidence Interval Using Known Population Standard Deviation

```python
# population standard deviation is 0.2 and known confidence interval is 0.99
known_std=0.2
confidence_interval_known=0.99
```

```python
alpha=1-confidence_interval_known
print(f"Alpha value is : {alpha:.4f}")
```

```python
std_error=known_std/math.sqrt(n)
print(f"Standard error is : {std_error:.4f}")
```

```python
z_critical_value = stats.norm.ppf(1-alpha/2)
print(f"Critical value is : {z_critical_value:.4f}")
```

```python
margin_error_known=z_critical_value*std_error
print(f"Marginal error is : {margin_error_known:.4f}")
```

```python
lower_bound_known=mean-margin_error_known
upper_bound_known=mean+margin_error_known
```

```python
print(f"The {confidence_interval_known*100}% confidence interval for Using Known Population Standard Deviation :- ")
print(f"lower bound : {lower_bound_known:.4f}")
print(f"upper Bound : {upper_bound_known:.4f}")
```

```python
print(f"confidence interval for the mean number of charachters printed before failure:\n{lower_bound_known:.4f}\n{upper_bound_known:.4f}")
```
