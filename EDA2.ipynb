# 2. Encoding Techniques:

```python
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
```

```python
ohe=OneHotEncoder()
le=LabelEncoder()
```

```python
#selecting categorical column for encoding
categorical_col=data.select_dtypes(include=["object"]).columns.tolist()
```

```python
ohe_data=pd.DataFrame()
le_data=data.copy()
```

```python
for column in categorical_col:
    num_categories=data[column].nunique()
    if num_categories<=5:
        # One-Hot encode for less than 5 categories
        encoded_feature=ohe.fit_transform(data[[column]])
        encoded_data=pd.DataFrame(encoded_feature.toarray(),columns=ohe.get_feature_names_out([column]))
        ohe_data=pd.concat([ohe_data,encoded_data],axis=1)
    else:
        # Label encode for more than 5 categories
        encoded_feature=le.fit_transform(data[column])
        le_data[column]=encoded_feature
```

One-Hot Encoding

* Pros:

  - No Ordinality: Avoids implying an order between categories.
  - Improved Accuracy: Can enhance model performance, especially for algorithms sensitive to feature scaling.

* Cons:

  - Dimensionality: Can significantly increase the number of features.
  - Sparsity: Leads to sparse data, which can be challenging for some algorithms.

Label Encoding

- Pros:

  - Simplicity: Easy to implement and doesn't increase dimensionality.
  - Suitable for Ordinal Data: Effective when a clear order exists between categories.

- Cons:

  - Artificial Order: Imposes an artificial order on nominal data.
  - Potential for Bias: The numerical values can mislead models sensitive to feature scaling.

**how outliers can affect model performance**

- **Bias:** Outliers can heavily influence a model, pulling the "best fit" line or decision boundary away from the true pattern of the majority of the data.
- **Overfitting:** The model might over-adapt to outliers, learning their specific nuances instead of generalizing well to unseen data.
- **Increased error:** Outliers often lead to larger errors during model training, as they deviate significantly from the expected values.
- **Skewed metrics:** Evaluation metrics, such as accuracy or R-squared, can be misleadingly influenced by the presence of outliers
