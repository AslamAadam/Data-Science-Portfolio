```python
data=data.fillna(data.mean())
data.info()
```

```python
data.describe()
```

```python
data.duplicated().value_counts()
```

```python
data=data.drop_duplicates()
```

```python
def outlier_detection():
  data_outliers=[]
  for column in data.columns:
    Q1=data[column].quantile(0.25)
    Q3=data[column].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    data_outlier=data[(data[column]<lower_bound)|(data[column]>upper_bound)].index
    data_outliers.extend(data_outlier)
  data_outliers=list(set(data_outliers))
  clean_data=data.drop(data_outliers)
  return clean_data
```

```python
outlier_detection()
```

```python
data.info()
```

```python
clean_data=outlier_detection()
```

```python
for column in data.columns:
  plt.figure(figsize=(12,8))
  plt.subplot(1,2,1)
  sns.boxplot(data[column])
  plt.title(f"{column} before")
  plt.subplot(1,2,2)
  sns.boxplot(clean_data[column])
  plt.title(f"{column} After")
plt.tight_layout()
plt.show()
```
