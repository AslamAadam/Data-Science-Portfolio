#Data Visualization

##Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns
for column in dt1_numerical.columns:
    plt.figure(figsize=(10,6))
    sns.histplot(dt1_numerical[column], kde=True,color='purple')
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.title(f"Histogram of {column}")
    plt.xticks(rotation=90)
    plt.show()
```

####After analyzing the histogram:

* **skewness:** is the degree of assymetry observed in a probability distribution.
   - Types of skewness:
    - Right skew (also called ***positive skew***). A right-skewed distribution is longer on the right side of its peak than on its left.
    - Left skew (also called ***negative skew***). A left-skewed distribution is longer on the left side of its peak than on its right.
    - Zero skew.

  - Here, the skewness is **positive skew**, except for histogram of discount rate.

* **Outliers:** an observation that lies an abnormal distance from other values in a random sample from a population.
  
  * Here outliers are present. They are further visible at boxplots.





##Boxplots

```python
for column in dt1_numerical.columns:
  plt.figure(figsize=(8,8))
  plt.boxplot(dt1_numerical[column])
  plt.xlabel(column)
  plt.title(f"Boxplot of {column}")
  plt.show()
```

####Afteranalyzing the boxplot:
* **The interquartile range:**

  - (IQR) is the difference between the first quartile (Q1) and the third quartile (Q3) of a dataset.
      
      IQR=Q3-Q1
  - It represents the spread of the middle 50% (median) of the data.
  - It is represented as a **yellow line** in each boxplot graphs.

* **Outlires:**
  
  - It is a value that lies in the extremes of a data series and thus can affect the overall observation.
  
  - The values that fall above or below the end of the whiskers, they are plotted as dots.

  - These dots are the outliers.

  Here we can comfirm the outlier presence.
  
* **Extreme value** are present in the dataset.

   - The extreme values (within 1.5 times the inter-quartile range from the upper or lower quartile) are the ends of the lines extending from the IQR. Points at a greater distance from the median than 1.5 times the IQR are plotted individually as asterisks. These points represent potential outliers.
* our dataset contain an unusual distribution most of the value are far away from the whiskers
* by analysing the graph we can conclude that most of the discount given is comparitively very less as it is left skewed

##Standardization of Numerical Variables

**Z-score normalization** refers to the process of normalizing every value in a dataset such that the mean of all of the values is 0 and the standard deviation is 1.

We use the following formula to perform a z-score normalization on every value in a dataset:

**z=x-mu/sigma**

where:
- z : new variable to store the result
- x : Original value
- μ or mu : Mean of data
- σ or sigma : Standard deviation of data

```python
Std_dt=(dt1_numerical-mean_values)/std_dev_values

plt.figure(figsize=(15,7))
plt.subplot(1,2,1)
sns.boxplot(dt1_numerical)
plt.title("Before Standardization")
plt.xticks(rotation=90)

plt.subplot(1,2,2)
sns.boxplot(Std_dt)
plt.title("After Standardization")
plt.xticks(rotation=90)
plt.tight_layout()
```

# Conversion of Categorical Data into Dummy Variables

```python
dt1_categorical.head(10)
```

Since most machine learning models only accept numerical variables, preprocessing the categorical variables becomes a necessary step. We need to convert these encode categorical
variables to numbers such that the model is able to understand and extract valuable information

* There are multiple ways for doing it
 - using one_hot_encoder
 - using pandas get_dummies
* both of the process give us binary result and creates seprate columns
  for each categorical values
* importantly one_hot_encoder have an unknown_value handler built in
 so it is most commonly used for getting dummies/ converting categorical value.


 ***In this context, it's recommended to use one-hot encoding for categorical columns.***

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(handle_unknown='ignore',dtype=int,sparse_output=False)
```

```python
dt1_categorical.dtypes
```

```python
dummy_df=pd.get_dummies(dt1_categorical,drop_first=True,dtype=int)
dummy_df
```

```python
ohe_dt=pd.DataFrame(ohe.fit_transform(dt1_categorical),columns=ohe.get_feature_names_out())
ohe_dt.head(10)
```

```python
final_dt=pd.concat([dt1_numerical,ohe_dt],axis=1)
final_dt.head()
```

```python
final_dt.to_csv('final_data.csv',index=False)
```
