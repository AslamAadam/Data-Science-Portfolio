```python
import pandas as pd
import numpy as np
import scipy.stats as stats
from scipy.stats import chi2_contingency
```

```python
data={"Satisfaction":['Very Satisfied','Satisfied','Neutral','Unsatisfied','Very Unsatisfied'],
      "Smart Thermostat":[50,80,60,30,20],"Smart Light":[70,100,90,50,50]}
```

```python
data=pd.DataFrame(data,columns=['Satisfaction','Smart Thermostat','Smart Light'])
data=data.set_index('Satisfaction')
data
```

# State the Hypotheses:

* **NULL HYPOTHESIS** (H0): There is no association between the type of smart home device purchased and the customer satisfaction level.

* **ALTERNATE HYPOTHESIS** (H1): There is an association between the type of smart home device purchased and the customer satisfaction level.

# Compute the Chi-Square Statistic:

```python
chi_stats,p,dof,expected=chi2_contingency(data)
```

```python
chi_stats
```

```python
p
```

```python
dof
```

```python
print("\nExpected:")
print(pd.DataFrame(expected, index=data.index, columns=data.columns))

print("\nActual:")
print(pd.DataFrame(data, index=data.index, columns=data.columns))
```

```python
alpha=0.05 #Using the significance level (alpha) of 0.05
```

# Determine the Critical Value:


```python
critical_value=stats.chi2.ppf(1-alpha,dof)
critical_value
```

# Make a Decision:

```python
if chi_stats>=critical_value:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

### by analysis the result we made the conclusion
 - fail to reject the NULL HYPOTHESIS (H0)
 - since it failed the reject NULL HYPOTHESIS
 we can say that :
 There is no association between the type of smart home device purchased and the customer satisfaction level.
