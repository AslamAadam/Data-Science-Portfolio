# Prime Numbers

```python
# Enter the input number
num=int (input("Enter the number :"))
# 0 and 1 are neither prime or composite number
if num==0 or num==1:
  print("The number is not a prime number")
  # If the number is greater than 1 we use for loop
if num>1:
  # we itrate the for loop with variable 'i' and it can be upto the given number
    for i in range(2,num):
      # if the number can be devided by i, then it is not a prime number
      if num%i==0:
        print("The number is not a prime number")
        break
        # if the number can not be devided by i, then it is a prime number
    else:
      print("The number is a prime number")
```

#Word Counter


```python
# Defining a class
def word_count():
  # Enter the string
  text=(input("Enter the input text :").lower())
  # Splitting the words while neglecting fullstops
  text=text.replace(".","").split()
  # Creating a dictionary
  wc={}
  # Printing the word count
  print("The number of words in the text is:",len(text))
  # Printing the number of words in the text
  print("The number of unique words in the text is:",len(wc))
  # Counting the words
  for word in text:
    # Checking if the word is already in the dictionary
    if word in wc:
      # If the word is already in the dictionary, then increment the count
      wc[word]+=1
      # If the word is not in the dictionary, then add it to the dictionary with a count of 1
    else:
      wc[word]=1
      # Printing the dictionary
  for key in list(wc.keys()):
    print(key, ":",wc[key])
    return wc
  # Calling the function
word_count()
```

# Check for palindrome

```python
def is_palindrome():
  # Recieving the input
  s=input("Enter the string :")
  # Checking if the string is a palindrome
  if s==s[::-1]:
    # If the string is a palindrome, then print the message
    print("The string is a palindrome")
    # If the string is not a palindrome, then print the message
  else:
    print("The string is not a palindrome")
    return is_palindrome()
  # Calling the function
is_palindrome()
```
