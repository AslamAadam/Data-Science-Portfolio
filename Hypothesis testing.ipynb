# State the Hypotheses statement:

* **NULL HYPOTHESIS (H0):**The weekly operating costs remains as the theoretical weekly operating cost model
* **ALTERNATE HYPOTHESIS (H1):**The weekly operating costs are higher than that of theoretical weekly operating cost model

# Calculate the Test Statistic:

```python
import math
```

```python
x_bar=3050
mu=1000+(5*600)
std=5*25
n=25
dof=n-1
```

```python
mu
```

```python
std
```

```python
dof
```

```python
se=std/math.sqrt(n)
se
```

```python
t_stat=(x_bar-mu)/se
t_stat
```

# Determine the Critical Value:

```python
import scipy.stats as stats
```

```python
t_crit=stats.t.ppf(1-0.05/2,dof) # alpha is 0.05
t_crit
```

# Make a Decision:

```python
if t_stat>t_crit:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
```
