```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel("Online retail.xlsx")
```

```python
data.head()
```

```python
transaction=data.iloc[:, 0].apply(lambda x: x.split(','))
```

```python
transaction=transaction.apply(lambda x:list(set(x)))
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
TE=TransactionEncoder()
```

```python
te_ary=TE.fit_transform(transaction)#.#transform(transaction)
```

```python
te_ary
```

```python
new_data=pd.DataFrame(te_ary,columns=TE.columns_,dtype=int)
```

```python
new_data
```

```python
from apyori import apriori
from mlxtend.frequent_patterns import association_rules , apriori
```

```python
frequent_itemsets=apriori(new_data,min_support=0.01,use_colnames=True)
```

```python
frequent_itemsets.head()
```

```python
rules=association_rules(frequent_itemsets,metric="lift",min_threshold=1)
```

```python
print(rules.head())
```

```python
# Interpretation of the results
for index, rule in rules.head(10).iterrows():
    antecedents = list(rule['antecedents'])
    consequents = list(rule['consequents'])
    print(f"Rule: {antecedents} -> {consequents}")
    print(f"Support: {rule['support']}")
    print(f"Confidence: {rule['confidence']}")
    print(f"Lift: {rule['lift']}\n")
```

# Interview Questions:

1.	What is lift and why is it important in Association rules?

- Lift quantifies the strength of association between items
- A lift greater than 1 indicates a positive association
- A lift less than 1 indicates a negative association
- Higher lift values indicate stronger associations

2.	What is support and Confidence. How do you calculate them?

- **Support** indicates how frequently an itemset appears in the dataset.
  - It's calculated by dividing the number of transactions containing the itemset by the total number of transactions.
- **Confidence** measures how often the consequent is found in transactions that contain the anteceden
  - It's calculated by dividing the support of the itemset {antecedent, consequent} by the support of the antecedent.

3.	What are some limitations or challenges of Association rules mining?

- Finding relevant rules in large datasets
- Handling rare items
- Dealing with spurious associations
- Choosing appropriate support and confidence thresholds

these are some the challenges of association rules mining
