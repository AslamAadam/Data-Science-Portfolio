#### Item-based filtering:

* This method focuses on the relationships between items. It identifies similar items based on how users have rated or interacted with them

## 2. What is collaborative filtering, and how does it work?

Collaborative filtering is a technique used by recommender systems to suggest items to users based on the preferences of similar users.

how it works:
- The system collects data on user interactions, such as ratings, purchases, or views.
- The system finds users with similar tastes by comparing their interaction data
- The system recommends items that similar users have liked or interacted with, but which the target user has not yet encountered.
